```python
import numpy as np
import pandas as pd
import statistics as st
```

```python
data = pd.read_csv('./MiningProcess_Flotation_Plant_Database.csv', index_col=False, decimal = ',')
```

```python
data.head()
```

```python
data.dtypes
```

```python
L = len(data)
```

```python
L
```

```python
FloatationColumnAirFlowMean = []
FloatationColumnAirFlowMedian = []
FloatationColumnLevelMean = []
FloatationColumnLevelMedian = []
for i in range(L):
    val_airflow = [data['Flotation Column 01 Air Flow'][i], data['Flotation Column 02 Air Flow'][i], data['Flotation Column 03 Air Flow'][i], data['Flotation Column 04 Air Flow'][i], data['Flotation Column 05 Air Flow'][i], data['Flotation Column 06 Air Flow'][i], data['Flotation Column 07 Air Flow'][i]]
    val_level = [data['Flotation Column 01 Level'][i], data['Flotation Column 02 Level'][i], data['Flotation Column 03 Level'][i], data['Flotation Column 04 Level'][i], data['Flotation Column 05 Level'][i], data['Flotation Column 06 Level'][i], data['Flotation Column 07 Level'][i]]
    FloatationColumnAirFlowMean.append(st.mean(val_airflow))
    FloatationColumnAirFlowMedian.append(st.median(val_airflow))
    FloatationColumnLevelMean.append(st.mean(val_level))
    FloatationColumnLevelMedian.append(st.median(val_level))
    if ((i % 100000) == 0):
        print(f"{i} entries out of {L-1} entries computed")
```

```python
data['Floatation Column Air Flow Mean'] = FloatationColumnAirFlowMean
data['Floatation Column Air Flow Median'] = FloatationColumnAirFlowMedian
data['Floatation Column Level Mean'] = FloatationColumnLevelMean
data['Floatation Column Level Median'] = FloatationColumnLevelMedian
```
